```python
# plotting a bar chart for Segments vs total Count

ax = sns.countplot(x = 'Segment',data = df, hue='Region')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for Region vs total Sales

sales_reg = df.groupby(['Region'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)

sns.barplot(x = 'Region',y= 'Sales' ,data = sales_reg)
```

```python
df.groupby(['Year'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)
```

```python
# plotting a bar chart for Years vs total Sales

sales_reg = df.groupby(['Year'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)

sns.barplot(x = 'Year',y= 'Sales' ,data = sales_reg)
```

```python
# plotting a bar chart for Country vs total Sales

sales_reg = df.groupby(['Country'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)

sns.barplot(x = 'Country',y= 'Sales' ,data = sales_reg)
```

```python
#State sales count
sns.set(rc={'figure.figsize':(60,30)})
ax = sns.countplot(data = df, x = 'State')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Year count of sales.
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.countplot(data = df, x = 'Year')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Top 10 sales by Year.
sales_state = df.groupby(['Year'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'Year',y= 'Sales')
```

```python
#Top 10 sales by Subcategory
sales_state = df.groupby(['SubCategory'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'SubCategory',y= 'Sales')
```

```python
#Top 10 sales by Category
sales_state = df.groupby(['Category'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'Category',y= 'Sales')
```

```python
#Top 10 sales by Region
sales_state = df.groupby(['Region'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = sales_state, x = 'Region', y= 'Sales')
```

```python
#Top 10 sales by City
sales_state = df.groupby(['City'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = sales_state, x = 'City', y= 'Sales')
```

```python
#Top 10 sales by state
sales_state = df.groupby(['State'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = sales_state, x = 'State', y= 'Sales')
```
